# Punto 1a

Punto 1a: Make plots of T(x,z) for 3 different ridge half-spreading rates: "slow" (0.5 cm/year), "moderate" (4 cm/year) and "fast" (10 cm/year). Use colours or contours to display the temperature field in 2-D. Assume a thermal diffiusivity value of 10-6 m2/s. Use a "reasonable" range of x, y values. How do these 3 plots differ?

```python
#Constantes
TsTo = 1300 #K
To = 273.15 #K
k = 10**-6 #Thermal Diffiusitivy (m^2/s)

#ARREGLO DE X y Y
x = np.linspace(-10000,10000,1000)
y = np.linspace(0,-15000,1000)
X,Y=np.meshgrid(x,y)
Q,W=np.meshgrid(x,y)
A,S=np.meshgrid(x,y)
```

```python
def temp(x,y,v):
    return (TsTo*erf(abs(y)/(2*np.sqrt(k*(abs(x)/v))))) + To
```

```python
v1 = 0.5/(100*365*24*3600)
T= temp(X,Y,v1)
plt.subplots(figsize=(20, 5))
imshow(T,extent=[x.min(), x.max(), y.min(), y.max()])
cb=plt.colorbar()
plt.title('Temperatura 0,5cm/yr de Velocidad', fontsize=20)
plt.xlabel('Distancia (m)', fontsize=15)
plt.ylabel('Profundidad (m)', fontsize=15)
```

```python
v2 = 4.0/(100*365*24*3600)
T2= temp(Q,W,v2)
plt.subplots(figsize=(20, 5))
imshow(T2,extent=[x.min(), x.max(), y.min(), y.max()])
cb=plt.colorbar()
plt.title('Temperatura 4cm/yr de Velocidad', fontsize=20)
plt.xlabel('Distancia (m)', fontsize=15)
plt.ylabel('Profundidad (m)', fontsize=15)
```

```python
v3 = 10.0/(100*365*24*3600)
T3 = temp(A,S,v3)
plt.subplots(figsize=(20, 5))
imshow(T3,extent=[x.min(), x.max(), y.min(), y.max()])
cb=plt.colorbar()
plt.title('Temperatura 10cm/yr de Velocidad', fontsize=20)
plt.xlabel('Distancia (m)', fontsize=15)
plt.ylabel('Profundidad (m)', fontsize=15)
```

#Punto 1b: Thermal boundary layer

Punto 1b: Now using the concept of a "thermal boundary layer" discussed in class, plot profiles showing the thickness of the thermal lithosphere as a function of x, for each of your 3 plates (slow/medium/fast).

```python
def profun(x,v):
    return -2.32*sqrt(k*(abs(x)/v))
```

```python
x = np.linspace(-10000,10000,1000)
fig = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
plt.subplot(1,3,1)
plot(x,profun(x,v1))
plt.title('Porfundidad con 0.5cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)

plt.subplot(1,3,2)
plot(x,profun(x,v2))
plt.ylim(-20000,0)
plt.title('Profundidad con 4cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)

plt.subplot(1,3,3)
plot(x,profun(x,v3))
plt.ylim(-20000,0)
plt.title('Profundidad con 10cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)

plt.show()
```

#Punto 1c: Oceanic bathymetry

Punto 1c: For each of the same 3 plates, calculate the ocean bathymetry (i.e. the surface topography of oceanic lithosphere), and plot it along with the plate thickness, to show a "cross section" of oceanic lithosphere as a function of x. in each case. Assume a thermal expansion coefficient α of 1.6x10-5 per °C and reasonable assumptions about densities.

```python
rm = 3300  #kg/m3
rw = 1000  #kg/m3
alpha = 1.6*10**-5 
TmTs = 1200

def batime(x,v):
    return (-rm/(rm-rw))*alpha*TmTs*2*(sqrt(k*abs(x)/v)/sqrt(pi))
```

```python
fig = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
plt.subplot(1,3,1)
plot(x,profun(x,v1), label= "thermal boundary", c='r')
plot(x,batime(x,v1), label= "Bathymetry", c='b')
plt.title('Corte para 0.5cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)
legend(fontsize=15, loc='lower center')

plt.subplot(1,3,2)
plot(x,profun(x,v2), label= "thermal boundary", c='r')
plot(x,batime(x,v2), label= "Bathymetry", c='b')
plt.title('Corte para 4cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)
legend(fontsize=15, loc='lower center')

plt.subplot(1,3,3)
plot(x,profun(x,v3), label= "thermal boundary", c='r')
plot(x,batime(x,v3), label= "Bathymetry", c='b')
plt.title('Corte para 10cm/yr de Velocidad', fontsize=12)
plt.xlabel('Distancia (m)', fontsize=10)
plt.ylabel('Profundidad (m)', fontsize=10)
legend(fontsize=15, loc='lower center')

plt.show()
```

#Punto 1d: Relaxation times  

Punto 1d: In class we showed how the characteristic relaxation times of a Maxwell viscoelastic material vary exponentially with temperature. Plot a profile of the characteristic relaxation time as a function of temperature, using an the appropriate range of temperatures for oceanic crust (you may wish to use a logarithmic scale for the τ axis).

```python
#Denino las constantes, Q,R,A,E
Q = 240*10**3 #Calor de activación
R = 8.3145 #Constante de los gases
A = 8*10**(-3) 
E = 50*10**(-3) #modulo de young
n = 3.1
```

```python
def tao(t,esfu):
    return ((esfu**(1-n))*(e**(Q/(R*t))))/(2*A*E)
```

```python
t = np.linspace(0,2000,100)
```

```python
figure(figsize(8,5))
#Aplicando de a 1kbar = 3,3Km y que 1kbar = 10^8Pa puedo obtener el esfuerzo
#A 10Km
plt.semilogy(t,tao(t,(10*(10**8)*(1/3.3))),label="10km")
#A 30Km
plt.semilogy(t,tao(t,(30*(10**8)*(1/3.3))),label="30km")
#A 60Km
plt.semilogy(t,tao(t,(60*(10**8)*(1/3.3))),label="60km")
#A 90Km
plt.semilogy(t,tao(t,(90*(10**8)*(1/3.3))),label="90km")
#A 120Km
plt.semilogy(t,tao(t,(120*(10**8)*(1/3.3))),label="120km")
legend(fontsize=12)
plt.grid(True)
plt.xticks(np.arange(0,2001,200))
plt.ylim(ymax=10**19)
plt.title('Tiempo de Relajacion en funcion de T', fontsize=15)
plt.xlabel('Temperatura (K)', fontsize=10)
plt.ylabel('Tiempo de Relajacion (s)', fontsize=10)
```

#Punto 1e: Elastic/viscous transition

Punto 1e: Using your plot from part (d), estimate an "elastic/viscous transition temperature", i.e. an approximate temperature for which the behaviour of your material changes from being "elastic" on geological timescales, to "viscous" (you may wish to consider a range of values of relaxation time τ, since the transition will not be instantaneous - however, you should notice that the transition is fairly sharp). Plot your chosen elastic/viscous transition temperature as a function of x for oceanic lithosphere using T(x, y) for oceanic plates. Plot these along with the lithospheric profile plots you produced in part (c). How does the "thermal thickness" of the lithosphere compare with the "elastic thickness" of the lithosphere?

```python
#De la figura anterior podemos observar que el cambio entre elastico/viscoso está al rededor de 10^10s - 10^12s
#Gracias al cambio de las propiedades mecanicas de las rocas, y podemos sacar la Temperatura promedio en ese cambio
l1=linspace(1,1,100)*10**10
l2=linspace(1,1,100)*10**12

figure(figsize(8,5))
#Aplicando de a 1kbar = 3,3Km y que 1kbar = 10^8Pa puedo obtener el esfuerzo
#A 10Km
plt.semilogy(t,tao(t,(10*(10**8)*(1/3.3))),label="10km")
#A 30Km
plt.semilogy(t,tao(t,(30*(10**8)*(1/3.3))),label="30km")
#A 60Km
plt.semilogy(t,tao(t,(60*(10**8)*(1/3.3))),label="60km")
#A 90Km
plt.semilogy(t,tao(t,(90*(10**8)*(1/3.3))),label="90km")
#A 120Km
plt.semilogy(t,tao(t,(120*(10**8)*(1/3.3))),label="120km")
legend(fontsize=12)
plt.grid(True)
plt.xticks(np.arange(0,2001,200))
plt.ylim(ymax=10**19)
plt.title('Tiempo de Relajacion en funcion de T', fontsize=15)
plt.xlabel('Temperatura (K)', fontsize=10)
plt.ylabel('Tiempo de Relajacion (s)', fontsize=10)
plot(t,l1)
plot(t,l2)
plt.show()
print ("podriamos decir que la temperatura en la transicion es entre 410K y 500K")
```

```python
TT = T2
for w in range(1000):
    for e in range(1000):
        if T2[w,e] < 410.0:
            TT[w,e] = 0.0
        elif T2[w,e] > 450.0:
            TT[w,e] = 0.0
        else:
            TT[w,e] = T2[w,e]

plt.subplots(figsize=(20, 5))
imshow(TT,extent=[x.min(), x.max(), y.min(), y.max()], cmap='seismic')
cb=plt.colorbar()
plt.title('Perfil litosferico, con transicion elasto/viscosa para V=4cm/yr', fontsize=20)
plt.xlabel('Distancia (m)', fontsize=15)
plt.ylabel('Profundidad (m)', fontsize=15)
plot(x,profun(x,v2), label= "thermal boundary", c='g')
plot(x,batime(x,v2), label= "Bathymetry", c='y')
plt.ylim(ymin=-7000)
legend(fontsize=15, loc='lower center')
```

Podemos ver que el Thermal Thickness de la litosfera es mucho mas grande que el elastic thickness, es decir que la mayor parte de la litosfera se emvuelve bajo la thermal thickness. Tambien se puede ver que el elastic thickness representa una transicion mas lineal, es decir se se representa como una linea practicamente recta. Mientras que el thermal thickness tiene un comportamiento mas parabolico, lo cual se debe a sus valores muy cercanos a cero en el centro de la dorsal y sus grandes valores al alejarse.

Nota: Para representar la transicion elasto/viscosa se utlizaron los valores de temperatura T(x,z) entre 410K y 500k, para el caso donde se tiene una V=4cm/yr. Para representar esa parte en la grafica se mostro el resto como si la T fuera igual a cero, sin embargo esto unicamente fue realizado para mostrar la transicion y no representa los valores reales de la temperatura fuera del area de transicion, eso ya lo hicimos en la parta 1a.
